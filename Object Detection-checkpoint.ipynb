# **Dependencies**

```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

# Loading the Model 

```python
model=torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
model
```

# Checking For Detection with Imported Image URL

```python
img = 'https://static1.bigstockphoto.com/3/1/4/large1500/413549600.jpg'
```

```python
results = model(img)
results.print()
```

```python
%matplotlib inline 
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
results.render()
```

# Object Detection with Imported Video

```python
cap = cv2.VideoCapture('cat.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('Object Detection', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):   
        break
cap.release()
cv2.destroyAllWindows()
```
